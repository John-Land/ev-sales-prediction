# Enterprise Value to Sales Ratio (EV/Sales) prediction

A traditional measure of whether a stock is expensive or cheap, is to use a valuation multiple. <br>
A valuation multiple relates the market value of an asset relative to a key statistic that is assumed to relate to that value. <br>
To be useful, that statistic – whether earnings, cash flow or some other measure – must bear a logical relationship to the market value observed. <br>
For more information about valuation multiples, refer to below link: <https://en.wikipedia.org/wiki/Valuation_using_multiples#Valuation_multiples>

## Enterprise Value to Revenue (Sales) Ratio (EV/Sales) and its fundamental drivers

One of the most widely used valuation multiples is the Enterprise Value to Sales (Revenue) multiple, short EV/Sales. <br>
This is the ratio between the Enterprise Value of a company, divided by the company’s Revenue. <br>
The Enterprise value is the total market value of the company’s equity and net debt. <br>

Based on Financial Theory, the value of any asset is derived from its future cash flow potential. <br>
In its simplest form, where we expect the cash flow (CF) to grow at the same rate forever (g), the present value (PV) of an asset is calculated with below perpetuity formula. <br><br>
$PV = \frac{CF * (1+g)}{(r-g)}$ <br>

We can use this formula to derive the fundamental drivers of the EV/Sales ratio.

First, we replace the present value (PV) with the enterprise value (EV), and replace the cash flow (CF), with the company’s free cash flow (FCFF). <br><br>
$EV = \frac{FCFF*(1 + g)}{(r - g)}$<br>

We can now break down the company’s free cash flow (FCFF) into its components. <br><br>
$EV = \frac{(EBITDA-Taxes-CAPEX-deltaWC)*(1 + g)}{(r - g)}$<br>

Re-writing the numerator relative to EBITDA gives us :<br><br>
$EV = \frac{EBITDA*(1-\frac{Taxes}{EBITDA}-\frac{CAPEX}{EBITDA}-\frac{deltaWC}{EBITDA})*(1 + g)}{(r - g)} = \frac{EBITDA*(1-\frac{Taxes}{EBITDA}- \frac{reinvestment}{EBITDA})*(1 + g)}{(r - g)}$<br>

Re-writing the EBITDA into Sales * EBITDA Margin gives us:<br><br>
$EV = \frac{Sales*EbitdaMargin*(1-\frac{Taxes}{EBITDA}- \frac{reinvestment}{EBITDA})*(1 + g)}{(r - g)}$<br>

Finally, dividing both sides by the Sales, gives us the EV to Sales ratio:<br><br>
$\frac{EV}{Sales} = \frac{EbitdaMargin*(1-\frac{Taxes}{EBITDA}- \frac{reinvestment}{EBITDA})*(1 + g)}{(r - g)}$<br><br>

This formula gives us some understanding of what we would expect the fundamental drivers of the EV/Sales ratio to be. <br>
The fundamental drivers of the EV/Sales ratio are
1. Higher expected growth rate g, all else being equal, should lead to a higher EV/Sales
2. Higher expected Ebitda Margin, all else being equal, should lead to a higher EV/Sales
3. Higher reinvestment needed, all else being equal, should lead to a lower EV/Sales
4. Higher Taxes, all else being equal, should lead to a lower EV/Sales
5. Higher risk, captured by the discount rate r, all else being equal, should result in a lower EV/Sales

Based on above derivation, we will use below fundamental variables which relate to points 1-5, in order to predict EV/Sales. <br>

| Financial KPIs used for prediction | Related Fundamental Driver | Comment |
|---|---|---|
| Revenue_Forecast_CAGR_10y | Growth | Average Revenue growth projections for the next 10 years by analysts. |
| Avg_EBITDA_Margin_Forecast_10y | Ebitda Margin | Average Ebitda Margin projections for the next 10 years by analysts. |
| Avg_Capex_Margin_5y | Re-investment requirement | Average Capex as % of Revenues paid in the past 5 years. Proxy for re-investment requirement. |
| Marginal_Tax_Rate | Taxes | Latest marginal tax rate paid. Proxy for future expected Tax expenses. |
| Net_Debt_perc_Market_Cap | Risk | The more leverage/debt the company has, the riskier it is for both equity and debt capital providers. |
| Sector | Risk | Some Sectors are riskier than others. The risk of a company is partially determined by the business it operates in. |
| Operating_Region | Risk | Some regions are riskier than others, due to political or operational risk. The risk of a company is partially determined by where it does business. |

# Data Pre-Processing

```python
#load packages
import numpy as np #linear algebra     
import pandas as pd #tabular data processing
#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt #data visualization
import seaborn as sns #data visualization
```

```python
#load company data
path = 'data/'
file = 'data.csv'

company_data_orig = pd.read_csv(path+file)
company_data_orig.head()
```

```python
#rename all columns

#get copy of original data
company_data = company_data_orig.copy()

#column name dictionary mapping
column_translation_dictionary = {'Name': 'Name',
                                 'Full Ticker': 'Full_Ticker',
                                 'Ticker': 'Ticker', 
                                 'Sector': 'Sector', 
                                 'Industry': 'Industry_GICS',
                                 'Market Cap': 'Market_Cap',
                                 'Net Debt': 'Net_Debt',
                                 'Enterprise Value (EV)': 'EV', 
                                 'Revenue': 'Revenue',  
                                 'Revenue Forecast CAGR (10y)': 'Revenue_Forecast_CAGR_10y',
                                 'Avg EBITDA Margin Forecast (10y)': 'Avg_EBITDA_Margin_Forecast_10y',
                                 'Avg Capital Expenditures Margin (5y)': 'Avg_Capex_Margin_5y',
                                 'Marginal Tax Rate': 'Marginal_Tax_Rate',
                                 'Operating Region': 'Operating_Region',
                                 'Operating Country': 'Operating_Country'
                                 }

#re-name columns
for column in company_data.columns:
  company_data = company_data.rename(columns={column: column_translation_dictionary[column]})

company_data.head()
```

```python
#ensure columns have the correct data type
company_data['Market_Cap'] = company_data['Market_Cap'].apply(pd.to_numeric, errors='coerce')
company_data['Net_Debt'] = company_data['Net_Debt'].apply(pd.to_numeric, errors='coerce')
company_data['EV'] = company_data['EV'].apply(pd.to_numeric, errors='coerce')
company_data['Revenue'] = company_data['Revenue'].apply(pd.to_numeric, errors='coerce')
company_data['Revenue_Forecast_CAGR_10y'] = company_data['Revenue_Forecast_CAGR_10y'].apply(pd.to_numeric, errors='coerce')
company_data['Avg_EBITDA_Margin_Forecast_10y'] = company_data['Avg_EBITDA_Margin_Forecast_10y'].apply(pd.to_numeric, errors='coerce')
company_data['Avg_Capex_Margin_5y'] = company_data['Avg_Capex_Margin_5y'].apply(pd.to_numeric, errors='coerce')
company_data['Marginal_Tax_Rate'] = company_data['Marginal_Tax_Rate'].apply(pd.to_numeric, errors='coerce')
```

```python
#add EV to Sales ratio and Net Debt % of market cap
company_data['Net_Debt_perc_Market_Cap'] = company_data['Net_Debt']/company_data['Market_Cap']
company_data['EV_TTM_Revenue'] = company_data['EV']/company_data['Revenue']

#ensure columns have the correct data type
company_data['EV_TTM_Revenue'] = company_data['EV_TTM_Revenue'].apply(pd.to_numeric, errors='coerce')
company_data['Net_Debt_perc_Market_Cap'] = company_data['Net_Debt_perc_Market_Cap'].apply(pd.to_numeric, errors='coerce')
```

### Outlier Removal

```python
#check for outliers with EllipticEnvelope (Guassian)
from sklearn.covariance import EllipticEnvelope

#fit roburst Gaussian to data
X = company_data[['EV_TTM_Revenue', 'Revenue_Forecast_CAGR_10y','Avg_EBITDA_Margin_Forecast_10y','Net_Debt_perc_Market_Cap','Avg_Capex_Margin_5y','Marginal_Tax_Rate']]
cov = EllipticEnvelope(random_state=0, contamination=0.1).fit(X)

# Get outliers. predict returns 1 for an inlier and -1 for an outlier
inliers = cov.predict(X) == 1
outliers = cov.predict(X) == -1

#outlier removal
current_num_samples = company_data.shape[0]
company_data_new = company_data[inliers]
new_num_samples = company_data_new.shape[0]
print(current_num_samples - new_num_samples, 'companies were removed due to outliers.')
print('The new data shape is', company_data_new.shape)
```

# Machine Learning

```python
#ML dependancies

#data preprocessing pipeline to preprocess data table
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
#from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

#Grid search cross validation to find best hyper parameters
from sklearn.model_selection import GridSearchCV

#models considered
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import LinearSVR
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import VotingRegressor

#evaluation
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

#measure training time
import time
```

## Pre-Processing Pipeline

```python
# check training and test data shape
print('X_train has shape', X_train.shape)
print('y_train has shape', y_train.shape)
print('X_test has shape', X_test.shape)
print('y_test has shape', y_test.shape)
```

```python
# define categorical columns for transformation
categorical_variables = ['Sector', 'Operating_Region',]

# define numerical columns for transformation
numerical_variables = list(X_train.drop(columns=categorical_variables).columns)

# define column transformation for data preprocessing pipeline
column_transformer = ColumnTransformer(transformers=[
                                                 ('categorical_columns', OneHotEncoder(handle_unknown='ignore'), categorical_variables),
                                                 ('numeric_columns', StandardScaler(), numerical_variables)])
```

## Model Training and Hyper Parameter Tuning

```python
#ridge regression

#define hyperparameter candidates to test with cross validation
grid_values = {'model__alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 200, 300, 400, 500]}

#define model pipeline
model_1 = Pipeline([('column_transformer', column_transformer),
                    ('model', Ridge(max_iter = 100000))])

# run cross validation to find best parameters for the model
grid_model_1 = GridSearchCV(model_1, param_grid = grid_values, cv=5, scoring = ['r2', 'neg_mean_absolute_error', 'neg_root_mean_squared_error'] , refit = 'r2', n_jobs=-1)
start = time.time()
grid_model_1 = grid_model_1.fit(X_train, y_train)
stop = time.time()

#results
results = grid_model_1.cv_results_
best_index = np.argmin(results['rank_test_r2'])
grid_model_1_mae = abs(results['mean_test_neg_mean_absolute_error'][best_index])
grid_model_1_rmse = abs(results['mean_test_neg_root_mean_squared_error'][best_index])

print(f"Training time: {round(stop - start,2)}s")
print('Grid best parameter: ', grid_model_1.best_params_)
print('Grid best r2: ', grid_model_1.best_score_.round(3))
print('Grid best mae: ', grid_model_1_mae.round(3))
print('Grid best rmse: ', grid_model_1_rmse.round(3))
```

```python
#lasso regression

#define hyperparameter candidates to test with cross validation
grid_values = {'model__alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 200, 300, 400, 500]}

#define model pipeline
model_2 = Pipeline([('column_transformer', column_transformer),
                    ('model', Lasso(max_iter = 100000))])

# run cross validation to find best parameters for the model
grid_model_2 = GridSearchCV(model_2, param_grid = grid_values, cv=5, scoring = ['r2', 'neg_mean_absolute_error', 'neg_root_mean_squared_error'] , refit = 'r2', n_jobs=-1)
start = time.time()
grid_model_2 = grid_model_2.fit(X_train, y_train)
stop = time.time()

#results
results = grid_model_2.cv_results_
best_index = np.argmin(results['rank_test_r2'])
grid_model_2_mae = abs(results['mean_test_neg_mean_absolute_error'][best_index])
grid_model_2_rmse = abs(results['mean_test_neg_root_mean_squared_error'][best_index])

print(f"Training time: {round(stop - start,2)}s")
print('Grid best parameter: ', grid_model_2.best_params_)
print('Grid best r2: ', grid_model_2.best_score_.round(3))
print('Grid best mae: ', grid_model_2_mae.round(3))
print('Grid best rmse: ', grid_model_2_rmse.round(3))
```
